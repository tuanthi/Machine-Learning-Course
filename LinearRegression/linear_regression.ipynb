```python
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt
import re
```

```python
# Load the data
X = []
Y = []
for line in open('data_1d.csv'):
    x, y = line.split(',')
    X.append(float(x))
    Y.append(float(y))
    
X = np.array(X)
Y = np.array(Y)
plt.scatter(X, Y, edgecolor='black')
plt.show()
```

```python
# Calculate a and b
denominator1 = X.dot(X) - X.mean()*X.sum()
a1 = (X.dot(Y) - Y.mean()*X.sum()) / denominator1
b1 = (Y.mean()*X.dot(X) - X.mean()*X.dot(Y)) / denominator1
Yhat = a1*X + b1

plt.scatter(X, Y, edgecolor='black')
plt.plot(X, Yhat, c='r')
plt.show()
```

```python
a2 = (X.dot(Y) - Y.sum()*X.mean()) / denominator1
Yhat = a2*X + b1

plt.scatter(X, Y, edgecolor='black')
plt.plot(X, Yhat, c='r')
plt.show()
```

```python
denominator2 = X.dot(X) - X.mean()**2
a2 = (X.dot(Y) - Y.mean()*X.mean()) / denominator2
b2 = (Y.mean()*X.dot(X) - X.mean()*X.dot(Y)) / denominator2
Yhat = a2*X + b2

plt.scatter(X, Y, edgecolor='black')
plt.plot(X, Yhat, c='r')
plt.show()
```

```python
# Calculate R-quare
d1 = Y - Yhat
d2 = Y - Y.mean()
r2 = 1 - d1.dot(d1) / d2.dot(d2)
print('R-square:', r2)
```

```python
# Moore's law
X = []
Y = []
non_decimal = re.compile(r'[^\d]+')

for line in open('moore.csv'):
    r = line.split('\t')
    x = int(non_decimal.sub('', r[2].split('[')[0]))
    y = int(non_decimal.sub('', r[1].split('[')[0]))
    X.append(x)
    Y.append(y)
    
X = np.array(X)
Y = np.array(Y)
plt.scatter(X, Y, edgecolor='black')
plt.show()

Y = np.log(Y)
plt.scatter(X, Y, edgecolor='black')
plt.show()

denominator1 = X.dot(X) - X.mean()*X.sum()
a1 = (X.dot(Y) - Y.mean()*X.sum()) / denominator1
b1 = (Y.mean()*X.dot(X) - X.mean()*X.dot(Y)) / denominator1
Yhat = a1*X + b1
plt.scatter(X, Y, edgecolor='black')
plt.plot(X, Yhat, c='r')
plt.show()

d1 = Y - Yhat
d2 = Y - Y.mean()
r2 = 1 - d1.dot(d1) / d2.dot(d2)
print('a:', a1, 'b:', b1, 'R-square:', r2)

print('Time to double:', np.log(2)/a1, 'years')
```
