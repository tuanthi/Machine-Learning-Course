```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
v = np.array([1/3, 1/3, 1/3])
A = np.array([[0.3, 0.6, 0.1],
              [0.5, 0.2, 0.3],
              [0.4, 0.1, 0.5]])
```

```python
dist = []
for i in range(25):
    vf = v.dot(A)
    dist.append(np.linalg.norm(v-vf))
    v = vf
plt.plot(range(25), dist)
plt.show()
```

```python
def rotate_img(img):
    _img = np.zeros(img.shape)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            _img[i, j] = img[img.shape[0] - j - 1, i]
    return _img
plt.imshow(rotate_img(number))
plt.show()
```

```python
def np_is_symmetric(matrix):
    return np.allclose(matrix, matrix.T)
```

```python
def manual_is_symmetric(matrix):
    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            if matrix[i, j] != matrix[j, i]: return False
    return True
```

```python
def color(x, y):
    if (x > 0) ^ (y >0): return 'red'
    else: return 'blue'

a = np.random.uniform(-1, 1, 2000)
b = np.random.uniform(-1, 1, 2000)

for i, j in zip(a, b):
    plt.scatter(i, j, c = color(i, j), edgecolor='black', alpha=0.3)
plt.show()
```
