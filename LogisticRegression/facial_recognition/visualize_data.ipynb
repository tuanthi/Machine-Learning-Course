```python
import numpy as np
import matplotlib.pyplot as plt
from utils import getBinaryData
```

```python
labels = ['Anger', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']
X, Y = getBinaryData()
```

```python
for i in range(7):
    x, y = X[Y == i], Y[Y == i]
    N = len(y)
    if N > 0:
        j = np.random.choice(N)
        plt.imshow(x[j].reshape(48,48), cmap='gray')
        plt.title(labels[y[j]])
        plt.show()
```
